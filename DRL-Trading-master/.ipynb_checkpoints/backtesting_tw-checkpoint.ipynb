```python
#%% import packages
import pandas as pd
import pandas_datareader as pdr
from datetime import datetime
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline

# backtesting
import pyfolio

# configuration
from config import config
```

```python
#%% var definitions
# basic configuration
df = pd.read_csv('data/tw0050_10.csv')   # raw target stock data  
unique_trade_date = df[(df.Date > config.VALID_START) & (df.Date <= config.VALID_END)].Date.unique()   # complete trading dates 
stock_tic = sorted(df.tic.unique())
df_trade_date = pd.DataFrame({"Date": unique_trade_date})   # df storing unique trading date
start_date = datetime.strptime(config.VALID_START, "%Y-%m-%d")   # start date for datareader
end_date = datetime.strptime(config.VALID_END, "%Y-%m-%d")   # end date for datareader
rebalance_window = 63   # time interval for retraining 
validation_window = 63    # time interval for validation
#print(unique_trade_date)
#print(df_trade_date)

# baselines
twii = pdr.DataReader("^TWII", "yahoo", start_date, end_date)
#print(twii.head())
```

```python
#%% function definitions
def get_daily_return(df, target):   # calculate daily return from total asset value (account)
    df['daily_return'] = df[target].pct_change(1)
    print('Sharpe: ', (252**0.5)*df['daily_return'].mean()/df['daily_return'].std())
    # mean * 252 / ((std)^2 252) ^ (1/2) = 252^(1/2) * mean / std (assume that the risk free return = 0)
    return df

def get_account_value(model_name):   # get account value (total asset) values during trading time interval
    df_account_value = pd.DataFrame()   # store account value during trading time interval
    for step in range(rebalance_window+validation_window, len(unique_trade_date), rebalance_window):#len(unique_trade_date)+1, rebalance_window):
        df_tmp = pd.read_csv('results/{}/account_value_trade_{}_{}.csv'.format(config.RESULT_DIR, model_name, step))
        df_account_value = df_account_value.append(df_tmp, ignore_index=True)
    df_account_value = pd.DataFrame({'account_value': df_account_value['0']})   # col "0" is the account value
    account_pct_chg = df_account_value.account_value.pct_change(1)   # percentage chanege (1 day return) for account value
    sharpe = (252**0.5)*account_pct_chg.mean()/account_pct_chg.std()
    '''
        Should notice when the trading starts so that the df_trade_date need to be adjusted.
    '''
    df_account_value = df_account_value.join(df_trade_date[rebalance_window:].reset_index(drop=True))
    return df_account_value

def backtest_return(df):
    strategy_return= df.copy()
    strategy_return['Date'] = pd.to_datetime(strategy_return['Date'])
    strategy_return.set_index('Date', drop=False, inplace=True)
    strategy_return.index = strategy_return.index.tz_localize('UTC')
    #del strategy_return['Date']
    '''
        pyfolio.create_full_tear_sheet needs return sequence to be pd.Series
    '''
    ts = pd.Series(strategy_return['daily_return'].values, index=strategy_return.index)
    return ts

def backtest_position(model_name):
    '''
        Example:
            index         'AAPL'         'MSFT'          cash
            2004-01-09    13939.3800     -14012.9930     711.5585
            2004-01-12    14492.6300     -14624.8700     27.1821
            2004-01-13    -13853.2800    13653.6400      -43.6375
    '''
    #df_stock_info = pd.read_csv("done_data.csv")   # provide price information
    df_allocations = pd.DataFrame()   # df to record allocations
    for step in range(rebalance_window+validation_window, len(unique_trade_date), rebalance_window):
        df_tmp = pd.read_csv("results/{}/allocations_trade_{}_{}.csv".format(config.RESULT_DIR, model_name, step))
        df_allocations = df_allocations.append(df_tmp, ignore_index=True)
    df_allocations.columns = stock_tic + ["cash"]
    return df_allocations



def backtest_transaction(df):
    '''
        Example:
            index                  amount   price    symbol
            2004-01-09 12:18:01    483      324.12   'AAPL'
            2004-01-09 12:18:01    122      83.10    'MSFT'
            2004-01-13 14:12:23    -75      340.43   'AAPL'
    '''
    df_transaction_info = pd.read_csv("")   # 
```
